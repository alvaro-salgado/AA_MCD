Se crearán dos grupos del conunto de datos, el primero contiene a los sujetos con identificador 1-12 y el siguiente 13-30

Dado que el total de los sujetos fue dividido en 2 conjuntos de datos *train y test* los grupos no contendrán todos los sujetos.

```python
df_1 = df[df['subject'] <= 12].copy()
df_2 = df[df['subject'] >= 13].copy()
```
